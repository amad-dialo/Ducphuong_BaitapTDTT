```python
class Cylinder():
    def __init__(self,height,radius,pi):
        self.height=height
        self.radius=radius
        self.pi=pi
    def dien_tich_xung_quanh(self):
        return 2*self.pi*self.radius**2+2*self.pi*self.radius*self.height
    def thể_tích(self):
        return self.pi*self.radius**2*self.height
height,radius=(float(i) for i in input().split())
a=Cylinder(height,radius,3.14)
print(f'dien tich xung quanh:{a.dien_tich_xung_quanh():.2f}')
print(f'the tich la:{a.thể_tích():.2f}')

```

```python
class Date():
    def __init__(self,ngày,tháng,năm):
        self.ngày=ngày
        self.tháng=tháng
        self.năm=năm
    def valid(self):
        if self.tháng in [1,3,5,7,8,10,12] and 1<=self.ngày<=31:
            return True
        elif self.tháng in [4,6,9,11] and 1<=self.ngày <= 30:
            return True
        elif self.tháng==2 and self.năm%4==0 and 1<=self.ngày<=29:
            return True
        elif self.tháng==2 and self.năm%4!=0 and 1<=self.ngày<=28:
            return True
        else:
            return False
    def next_day(self):
        if not self.valid():
            return 'INVALID'
        else:
            if self.ngày==31:
                if self.tháng==12:
                    return f'1/1/{self.năm+1}'
                elif self.tháng in [1,3,5,7,8,10]:
                    return f'1/{self.tháng+1}/{self.năm}'
            elif self.tháng in [4,6,9,11] and self.ngày==30:
                return f'1/{self.tháng+1}/{self.năm}'
            elif self.tháng==2:
                if self.năm%4==0 and self.ngày==29:
                    return f'1/{self.tháng+1}/{self.năm}'
                elif self.năm%4!=0 and self.ngày==28:
                     return f'1/{self.tháng+1}/{self.năm}'
            else:
                return f'{self.ngày+1}/{self.tháng}/{self.năm}'

ngày,tháng,năm=(int(i) for i in input().split())
a=Date(ngày,tháng,năm)
a.next_day()

            
```

```python
class Point():
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def zero(self):
        if self.x==self.y==0:
            return 'M là gốc tọa độ'
        elif self.x==0 and self.y!=0:
            return 'M thuộc trục tung'
        elif self.y==0 and self.x!=0:
            return 'M thuộc trục hoành'
        else:
            return 'M không thuộc trung tung lẫn trục hoành'
    def distance(self):
        return (self.x**2+self.y**2)**(1/2)
    
x,y=(int(i) for i in input().split())
a=Point(x,y)
print(a.zero())
print(f'khoảng cách từ M đến gốc tọa độ là:{a.distance():.2f}')
```

```python
import os
def clear():
    os.system('cls' if os.name=='nt' else 'clear')
def pause():
    input('\nNhấn enter để tiếp tục....')
class ShoppingCart():
    def __init__(self):
        self.dic={}
    def add_items(self):
        clear()
        self.name=input('Tên sản phẩm:')
        self.price=int(input('Giá sản phẩm:'))
        self.dic[self.name]={'Giá':self.price}
        pause()
    def delete_item(self):
        clear()
        item=input('Nhập tên sản phẩm:')
        if item in self.dic:
            self.dic.pop(item)
            print('Đã xóa sản phẩm')
        else:
            print('Không tìm thấy sản phẩm ')
        pause()
    def check_list(self):
        clear()
        if not self.dic:
            print('Rỗng')
        else:
            print('Không rỗng')
        pause()
    def total(self):
        clear()
        total=0
        for value in self.dic.values():
            total+=value['Giá']
        print(total)
        pause()
    def show_dic(self):
        clear()
        print(f'{"Tên sản phẩm":<10}{"Giá"}')
        for items,values in self.dic.items():
            print(f'{items:<10}{values['Giá']}')
        pause()
    def delete_all(self):
        clear()
        self.dic.clear()
        print('Đã xóa giỏ hàng')
        pause()

a=ShoppingCart()
while True:
    clear()
    print('1.Thêm sản phẩm')
    print('2.Xóa sản phẩm')
    print('3.Kiểm tra rỏ hàng') 
    print('4.Tổng tiền')
    print('5.Xem giỏ hàng')
    print('6.Xóa giỏ hàng')
    print('7.Thoát chương trình') 
    try:
       choice=int(input('Nhập lựa chọn của bạn:'))
    except ValueError:
        print('Vui lòng nhập số')

    if choice<1 or choice>7:
        print('Vui lòng nhập lựa chọn hợp lệ')
    else:
        if choice==1:
            a.add_items()
        elif choice==2:
            a.delete_item()
        elif choice==3:
            a.check_list()
        elif choice==4:
            a.total()  
        elif choice==5:
            a.show_dic()
        elif choice==6:
            a.delete_all()
        elif choice==7:
            break
    
    
```

```python
class Rectangle():
    def __init__(self,w,h,k):
        self.w=w
        self.h=h
        self.k=k
    def area(self):
        return self.w*self.h*self.k**2
    def perimeter(self):
        return 2*self.k*(self.w+self.h)
    
w,h,k=(int(i) for i in input().split())
a=Rectangle(w,h,k)
dtich=a.area()
chu_vi=a.perimeter()
print(f'{dtich:.2f} {chu_vi:.2f}')
```

```python
class Fraction():
    def __init__(self,num,den):
        self.num = num
        self.den = den
        if den==0:
            raise ValueError('Mẫu số phải khác 0')
        if den<0:
            num=-num
            den=-den
        uoc=[]
        for i in range(1,min(abs(num),abs(den))+1):
            if num%i==0 and den%i==0:
                uoc.append(i)
        gcd=max(uoc)
        self.num//=gcd
        self.den//=gcd
    def add(self, other):
        num=self.num*other.den+self.den*other.num
        den=self.den*other.den
        return Fraction(num,den)
    def minus(self,other):
        num=self.num*other.den-self.den*other.num
        den=self.den*other.den
        return Fraction(num,den)
    def mul(self,other):
        num=self.num*other.num
        den=self.den*other.den
        return Fraction(num,den)
    def div(self,other):
        num=self.num*other.den
        den=self.den*other.num
        return Fraction(num,den)
    def __str__(self):
        return f'{self.num}/{self.den}'
a,b,op,c,d=input().split()
a,b,c,d=int(a),int(b),int(c),int(d)
f1=Fraction(a,b)
f2=Fraction(c,d)
if op=='+':
    result=f1.add(f2)
elif op=='-':
    result=f1.minus(f2)
elif op=='*':
    result=f1.mul(f2)
elif op=='/':
    result=f1.div(f2)
print(result)


        
        
   
            
```

```python
class BankAcount():
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=float(balance)
    def deposit(self,x):
        self.balance+=x
    def withdraw(self,x):
        if x<=self.balance:
            self.balance-=x
    def __str__(self):
        return f'{self.owner}:{self.balance:.2f}'
owner,balance=input().split()
acc=BankAcount(owner,balance)
n=int(input())
for i in range(n):
    cmd,x=input().split()
    x=float(x)
    if cmd=='DEPOSIT':
        acc.deposit(x)
    elif cmd=='WITHDRAW':
        acc.withdraw(x)
print(acc)


        
```

```python
import math
class Point3D:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def vector(self,other):
        return Vector3D(
            other.x-self.x,
            other.y-self.y,
            other.z-self.z)
class Vector3D:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def vo_huong(self,other):
        return Vector3D(
            self.y*other.z-self.z*other.y,
            self.z*other.x-self.x*other.z,
            self.x*other.y-self.y*other.x)
    def tich_vo_huong(self,other):
        return self.x*other.x+self.y*other.y+self.z*other.z
    def do_dai(self):
        return math.sqrt(self.x**2+self.y**2+self.z**2)
    def __str__(self):
        return f"Vector({self.x}, {self.y}, {self.z})"

A=Point3D(*map(float,input('Point A:').split()))
B=Point3D(*map(float,input('Point B:').split()))
C=Point3D(*map(float,input('Point C:').split()))
D=Point3D(*map(float,input('Point D:').split()))
AB=A.vector(B)
AC=A.vector(C)
BC=B.vector(C)
BD=B.vector(D)
n1=AB.vo_huong(AC)
n2=BC.vo_huong(BD)
cos=abs(n1.tich_vo_huong(n2))/(n1.do_dai()*n2.do_dai())
PHI=math.degrees(math.acos(cos))
print(f'{PHI:.2f}')
    
```

```python
class ComplexNumber():
    def __init__(self,realPart,imaginaryPart):
        self.realPart=realPart
        self.imaginaryPart=imaginaryPart
    def ComplexNumber_multiply(self,other):
        new_realPart=self.realPart*other.realPart-self.imaginaryPart*other.imaginaryPart
        new_imaginaryPart=self.realPart*other.imaginaryPart+self.imaginaryPart*other.realPart
        return ComplexNumber(new_realPart,new_imaginaryPart)
    def __str__(self):
        return f'{int(self.realPart)} {int(self.imaginaryPart)}'

r1,i1=map(float,input().split())
c1=ComplexNumber(r1,i1)
r2,i2=map(float,input().split())
c2=ComplexNumber(r2,i2)
ans=c1.ComplexNumber_multiply(c2)
print(ans)

```

```python
import math
class Triangle():
    def __init__(self,edge1,edge2,edge3):
        self.edge1=edge1
        self.edge2=edge2
        self.edge3=edge3
    def area(self):
        p=(self.edge1+self.edge2+self.edge3)/2
        area=math.sqrt(p*(p-self.edge1)*(p-self.edge2)*(p-self.edge3))
        return f'{area:.2f}'
edge1,edge2,edge3=(int(i) for i in input().split())
t=Triangle(edge1,edge2,edge3)
print(t.area())
```
