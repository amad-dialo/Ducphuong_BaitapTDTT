```python
n=int(input())
uoc=[]
def perf_num():
    for i in range(1,n):
        if n%i==0:
            uoc.append(i)
    if sum(uoc)==n:
        print(True)
    else:
        print(False)
perf_num()
```

```python
def position():
    lst=list(map(int,input().split()))
    k=int(input())
    for i in range(len(lst)):
        if lst[i]==k:
          return i+1
print(position())
```

```python
n=int(input())
def giai_thua(n):
    if n<=1:
        return 1
    else:
        return n*giai_thua(n-1)
print(giai_thua(n))
```

WW5A7

```python
n=input()
if '+' in n:
    a,b=n.split('+')
    print(int(a)+int(b))
elif '-' in n:
    a,b=n.split('-')
    print(int(a)-int(b))
elif '*' in n:
    a,b=n.split('*')
    print(int(a)*int(b))
elif '/' in n:
    a,b=n.split('/')
    c=int(a)/int(b)
    print(f'{c:.2f}')
```

```python
def pattern(n):
    dic={}
    lst=[]
    idx=0
    for i in n:
        if i not in dic:
            dic[i]=idx
            idx+=1
        lst.append(dic[i])
    return lst
a,b=input().split()
print(pattern(a)==pattern(b))
```
