```python

def binary_search(arr,k,left,right):
    if left>right:
        return -1
    mid=(left+right)//2
    if arr[mid]==k:
        return mid+1
    elif arr[mid]>k:
        return binary_search(arr,k,left,mid-1)
    elif arr[mid]<k:
        return binary_search(arr,k,mid+1,right)

arr=sorted(set(map(int,input().split())))
k=int(input())
print(binary_search(arr,k,0,len(arr)-1))
```

```python
def bubble_sort():
    lst=list(map(int,input().split()))
    for i in range(len(lst)-1):
           for j in range(len(lst)-i-1):
             if lst[j]>lst[j+1]:
              lst[j],lst[j+1]=lst[j+1],lst[j]
    print(lst)
def selection_sort():
    lstt=list(map(int,input().split()))
    arr=[]
    while lstt:
        m=min(lstt)
        lstt.remove(m)
        arr.append(m)
    print(arr)

bubble_sort()
selection_sort()





```

```python
lst=list(map(int,input().split()))
arr=[]
for x in lst:
    left=0
    right=len(arr)-1
    pos=len(arr)
    while left<=right:
        mid=(left+right)//2
        if arr[mid]>=x:
            pos=mid
            right=mid-1
        else:
            left=mid+1
    if pos==len(arr):
        arr.append(x)
    else:
        arr[pos]=x
print(len(arr))
```
