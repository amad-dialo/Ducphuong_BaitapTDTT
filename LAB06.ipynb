W6A9

```python
key1=input().split()
value1=list(map(int,(input().split())))
dic1=dict(zip(key1,value1))

key2=input().split()
value2=list(map(int,(input().split())))
dic2=dict(zip(key2,value2))

dic3={}
for k,v in dic1.items():
    if k not in dic3:
        dic3[k]=dic3.get(k,0)+v
for m,n in dic2.items():
    if m not in dic3:
        dic3[m]=dic3.get(m,0)+n
    else:
        dic3[m]=dic3[m]+n
dic3=dict(sorted(dic3.items()))
print(dic3)
```

```python
def similar():
    lstt=list(map(int,input().split()))
    k=int(input())
    for i in range(len(lstt)):
        if lstt[i]==k:
            return i+1
    else:
        return -1
print(similar())
```
