```python
def tong_binh_phuong(n):
    tong=0
    for i in str(n):
        tong+=int(i)**2
    return tong
def happy_number(n):
    passed=set()
    while n!=1:
      if n in passed:
        return False
      passed.add(n)
      n=tong_binh_phuong(n)
    return True
try:
   n=int(input())
   if happy_number(n):
      print('Yes')
   else:
      print('No')
except ValueError:
   print('Vui lonhf nhập số hợp lệ')
      

```
